#Removing Outliers method- 1: IQR (Inter Quartile Range)

```python
sns.boxplot(x = "total_deaths_per_million", data = ds)
```

```python
sns.displot(ds["total_deaths_per_million"])
plt.show()
```

```python
q1 = ds["total_deaths_per_million"].quantile(0.25)
q3 = ds["total_deaths_per_million"].quantile(0.75)
```

```python
q1, q3
```

```python
IQR = q3 - q1
```

```python
min_range = q1 - (1.5 * IQR)
max_range = q3 + (1.5 * IQR)
```

```python
min_range, max_range
```

```python
ds.shape
```

```python
new_ds = ds[ds["total_deaths_per_million"]< = max_range ]
```

```python
new_ds.shape
```

```python
plt.figure(figsize = (25, 5)
sns.boxplot(x ="total_deaths_per_million", data = ds)
plt.show()
```

```python
plt.figure(figsize =(25,5))
sns.boxplot(x ="total_deaths_per_million", data= new_ds)
plt.show()
```

#Removing Outliers method- 2 (Direct Method)

```python
min_range = ds["total_deaths_per_million"].mean() - (3* ds["total_deaths_per_million"].std())
```

```python
max_range = ds["total_deaths_per_million"].mean() + (3*ds["total_deaths_per_million"].std())
```

```python
min_range, max_range
```

```python
new_ds2 = ds[ds["total_deaths_per_million"] <= max_range]
```

```python
new_ds2
```

```python
sns.boxplot(x = "total_deaths_per_million", data = new_ds2)
plt.show()
```

#Removing Outliers method - 3 (Z Score)

```python
z_score = (ds["total_deaths_per_million"] - ds["total_deaths_per_million"].mean()) / (ds["total_deaths_per_million"].std())
```

```python
z_score
```

```python
ds["z_score"] = z_score
```

```python
new_ds3= ds[ds["z_score"]<3]
```
